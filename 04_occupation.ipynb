### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user'
data = pd.read_csv(url, sep='|')
data
```

### Step 3. Assign it to a variable called users.

```python
users = data
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
data['gender_binary'] = data['gender'].map({'M': 1, 'F': 0})
result = data.groupby('occupation')['gender_binary'].mean()
sort = result.sort_values(ascending=False)
print(sort)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min = data.groupby('occupation')['age'].min()
max = data.groupby('occupation')['age'].max()
print('min age',min)
print('max age',max)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
data.groupby(['occupation', 'gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
occupation_stats = data.groupby('occupation')['gender'].value_counts(normalize=True).unstack() * 100
occupation_stats = occupation_stats.fillna(0)
print(occupation_stats)
```
